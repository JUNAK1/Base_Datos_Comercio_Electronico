##Histograma

```python
try:
    mydb = connection.connect(host="containers-us-west-122.railway.app", database = "railway",user="root", passwd="nhbRgTlRs5H6CF9GEj4a",use_pure=True,port = 6798)
    sql =  "select Precio from Productos;"
    df = pd.read_sql(sql,mydb)
    mydb.close() #close the connection
except Exception as e:
    mydb.close()  
```

```python
df = df.to_numpy()
df = df.flatten()
print(df)
```

```python
mean = df.mean()
std = df.std()
x = np.linspace(df.min(), df.max(), 300)
y = 1/(std * np.sqrt(2 * np.pi)) * np.exp(-(x - mean)**2 / (2 * std**2))
```

```python
plt.plot(x, y)
plt.title('Distribución normal de precios de producto')
plt.xlabel('Precio')
plt.ylabel('Densidad')
plt.show()
```

##linear regrassion

```python
from scipy import stats
```

```python
slope, intercept, r, p, std_err = stats.linregress(fecha, pedido)

def myfunc(fecha):
  return slope * fecha + intercept

mymodel = list(map(myfunc, fecha))

plt.scatter(fecha, pedido)
plt.plot(fecha, mymodel)
plt.show()
```
